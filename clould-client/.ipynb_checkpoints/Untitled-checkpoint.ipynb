```python
import pandas as pd
```

```python
df = pd.read_excel("contact.xlsx", sheet_name=0) 
```

```python
df = df.iloc[:, :-6]
```

```python
print(df.head())
```

```python
df = df.iloc[1:].reset_index(drop=True)
```

```python
print(df.head())
```

```python
contacts = {}
```

```python
import re
for index, row in df.iterrows():
    # Extract phone number (assumed in third column)
    raw_number = str(row[2])
    digits = re.sub(r'\D', '', raw_number)  # Remove non-digit characters
    if len(digits) >= 10:
        phone = digits[-10:]  # Last 10 digits
    else:
        continue  # Skip if invalid number

    # Try to get name from column 1 or 2 (remove weird characters and whitespace)
    name = str(row[0]).strip().replace("\n", "").replace(".", "")
    name = name if name and name != phone else None

    # Set label to name or number
    contacts[phone] = name if name else phone
```

```python
contacts
```

```python
contacts["8778388464"] = "8778388464"
```

```python
contacts
```

```python
for i in contacts:
    if '?' in contacts[i]:
        contacts[i] = contacts[i].replace("?","")
        print(True)
```

```python
import json
with open("contacts.json", "w") as f:
    json.dump(contacts, f, indent=4)
```
